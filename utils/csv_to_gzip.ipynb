### IMPORTANDO LIB

```python
import pandas as pd
```

```python
CSV_INPUT_FILE = './data/votacao_candidato_munzona_2022_BRASIL.csv'
GZIP_OUTPUT_FILE = './data/votacao.parquet.gzip'
```

### LER CSV

```python
df = pd.read_csv(filepath_or_buffer=CSV_INPUT_FILE,sep=';',encoding='ISO-8859-1',usecols=['NR_TURNO','DS_ELEICAO',
                                                                    'TP_ABRANGENCIA','SG_UF','NM_MUNICIPIO','NR_ZONA',
                                                                    'DS_CARGO','NR_CANDIDATO','NM_CANDIDATO','NM_URNA_CANDIDATO',
                                                                    'DS_SITUACAO_CANDIDATURA','NR_PARTIDO','SG_PARTIDO','NM_PARTIDO',
                                                                    'NM_COLIGACAO','DS_COMPOSICAO_COLIGACAO','ST_VOTO_EM_TRANSITO','QT_VOTOS_NOMINAIS',
                                                                    'NM_TIPO_DESTINACAO_VOTOS','QT_VOTOS_NOMINAIS_VALIDOS','DS_SIT_TOT_TURNO'])
```

### SALVAR EM GZIP

```python
df.to_parquet(path=GZIP_OUTPUT_FILE,compression='gzip')
```
